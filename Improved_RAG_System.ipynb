```python
import os
import pytesseract
import re
import numpy as np
from unstructured.partition.pdf import partition_pdf
from langchain.text_splitter import RecursiveCharacterTextSplitter
from sentence_transformers import SentenceTransformer
import faiss
from rank_bm25 import BM25Okapi
from nltk.tokenize import word_tokenize
import nltk
from transformers import CLIPProcessor, CLIPModel
from PIL import Image
from openai import OpenAI
import pickle
import hashlib
from concurrent.futures import ThreadPoolExecutor


# Download NLTK data
nltk.download('punkt_tab')
nltk.download('punkt')

# Initialize models
model = SentenceTransformer('BAAI/bge-large-en-v1.5')
clip_model = CLIPModel.from_pretrained("openai/clip-vit-base-patch32")
clip_processor = CLIPProcessor.from_pretrained("openai/clip-vit-base-patch32")
client = OpenAI(base_url="https://openrouter.ai/api/v1", api_key="####")
```

```python


    ### Step 1: Extract Text and Image Metadata
    def extract_content_from_pdf(pdf_path):
        os.makedirs("figures", exist_ok=True)
        elements = partition_pdf(
            filename=pdf_path,
            strategy="hi_res",
            extract_images_in_pdf=True,
            infer_table_structure=True,
            extract_image_block_output_dir="figures"
        )
        content = ""
        image_metadata = {}
        image_counter = 1

        for element in elements:
            if element.category == "NarrativeText" or element.category == "Title":
                content += element.text + "\n"
            elif element.category == "Table":
                content += f"[Table {image_counter}]: {element.text}\n"
                image_counter += 1
            elif element.category == "Image":
                img_path = element.metadata.image_path if hasattr(element.metadata, 'image_path') else f"figures/figure-{image_counter}.jpg"
                # Use element.text if OCR'd, otherwise apply pytesseract
                description = element.text if element.text else pytesseract.image_to_string(Image.open(img_path))
                if not description.strip():
                    description = "No OCR text available"
                img_key = f"Image {image_counter}"
                content += f"[{img_key}]\n"
                image_metadata[img_key] = {"path": img_path, "description": description.strip()}
                image_counter += 1
            else:
                content += str(element.text) + "\n"
        return content.strip(), image_metadata



    ### Step 2: Chunking
    def recursive_chunking(text, chunk_size=200, overlap=100):
        text_splitter = RecursiveCharacterTextSplitter(chunk_size=chunk_size, chunk_overlap=overlap)
        return text_splitter.split_text(text)





    def store_embeddings_faiss(embeddings):
        embedding_dim = len(embeddings[0])
        index = faiss.IndexFlatL2(embedding_dim)
        np_embeddings = np.array(embeddings).astype('float32')
        index.add(np_embeddings)
        return index



    ### Step 4: Hybrid Search with Image Support
    def hybrid_search(query, index, chunks, image_metadata, alpha=0.5, top_k=5):
        image_ref = re.search(r"Image (\d+)", query, re.IGNORECASE)
        img_key = f"Image {image_ref.group(1)}" if image_ref else None
        search_chunks = [c for c in chunks if img_key in c] or chunks if img_key else chunks

        query_embedding = model.encode(query, normalize_embeddings=True)
        query_embedding = np.array([query_embedding]).astype('float32')
        distances, indices = index.search(query_embedding, len(chunks))
        vector_similarities = [1 / (1 + d) for d in distances[0]]

        tokenized_chunks = [word_tokenize(chunk.lower()) for chunk in chunks]
        bm25 = BM25Okapi(tokenized_chunks)
        tokenized_query = word_tokenize(query.lower())
        bm25_scores = bm25.get_scores(tokenized_query)

        combined_scores = []
        for i in range(len(chunks)):
            combined = alpha * bm25_scores[i] + (1 - alpha) * vector_similarities[i]
            combined_scores.append((i, combined))
        combined_scores.sort(key=lambda x: x[1], reverse=True)

        top_indices = [idx for idx, score in combined_scores[:top_k]]
        retrieved_chunks = [chunks[i] for i in top_indices]

        if img_key and img_key in image_metadata:
            retrieved_chunks.append(f"[Image Metadata]: {image_metadata[img_key]['description']}")

        return retrieved_chunks

    ### Step 5: Image Explanation with CLIP
    def explain_image_with_clip(query, image_path, img_key, image_metadata):
        image = Image.open(image_path).convert("RGB")
        inputs = clip_processor(text=[query], images=image, return_tensors="pt", padding=True)
        outputs = clip_model(**inputs)
        logits_per_image = outputs.logits_per_image
        return ("Image matches query" if logits_per_image.item() > 0.5 else "Image not clearly related") + f". OCR text: {image_metadata[img_key]['description']}"

    ### Step 6: Query RAG with Image Support
    def query_rag(user_query, index, chunks, image_metadata):
        retrieved_chunks = hybrid_search(user_query, index, chunks, image_metadata)

        image_ref = re.search(r"Image (\d+)", user_query, re.IGNORECASE)
        if image_ref:
            img_key = f"Image {image_ref.group(1)}"
            if img_key in image_metadata:
                img_path = image_metadata[img_key]["path"]
                image_explanation = explain_image_with_clip(user_query, img_path, img_key, image_metadata)
                retrieved_chunks.append(f"[Image Explanation]: {image_explanation}")

        prompt = f"""
        User Query: {user_query}
        Retrieved Context: {retrieved_chunks}
        You are an expert and intelligent assistant capable of logical thinking and critical thinking , your job is to provide the user with your thought process for solving the required problem given to you by the user.
        """

        response = client.chat.completions.create(
            model="deepseek/deepseek-r1:free",
            messages=[{"role": "user", "content": prompt}],
            max_tokens=1600,
            temperature=0.7,
            top_p=0.7,
            presence_penalty=0.2,
            frequency_penalty=0.2
        )
        reasoning = response.choices[0].message.content

        full_prompt = f"""
        User Query: {user_query}
        Chat History: {chat_history[-10:]}
        Retrieved Context: {retrieved_chunks}
        Reasoning: {reasoning}
        Based on the User Query, Retrieved Context and Reasoning by the previous model , Respond in the most humanly way as if your are explaining something be precise and Derive your answer for the given user query based on the Retrieved Chunks and Reasoning relevant to the question.
        """
        response2 = client.chat.completions.create(
            model="meta-llama/llama-4-maverick:free",
            messages=[{"role": "user", "content": full_prompt}],
            max_tokens=1600,
            temperature=0.7,
            top_p=0.7,
            presence_penalty=0.2,
            frequency_penalty=0.2
        )
        chat_history.append({"query": user_query, "response": response2.choices[0].message.content})
        return response2.choices[0].message.content


    def get_pdf_hash(pdf_path):
      with open(pdf_path, "rb") as f:
        return hashlib.md5(f.read()).hexdigest()

    def load_or_process_pdf(pdf_path):
        cache_dir = "cache"
        os.makedirs(cache_dir, exist_ok=True)
        pdf_hash = get_pdf_hash(pdf_path)
        cache_file = f"{cache_dir}/{pdf_hash}.pkl"

        if os.path.exists(cache_file):
            with open(cache_file, "rb") as f:
                return pickle.load(f)

        text, image_metadata = extract_content_from_pdf(pdf_path)
        chunks = recursive_chunking(text)


        def parallel_embedding(chunks):
            with ThreadPoolExecutor() as executor:
                return list(executor.map(lambda c: model.encode(c, normalize_embeddings=True).tolist(), chunks))
        embeddings = parallel_embedding(chunks)
        index = store_embeddings_faiss(embeddings)

        with open(cache_file, "wb") as f:
            pickle.dump((text, image_metadata, chunks, embeddings, index), f)
        return text, image_metadata, chunks, embeddings, index


```
